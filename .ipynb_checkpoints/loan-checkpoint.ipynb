```python
def out_outliers(data,Datos):
    valor_medio = data[Datos].mean()
    std = data[Datos].std()
    minimo = valor_medio - 3*std
    maximo = valor_medio + 3*std
    mascara_outliers = np.logical_or(data[Datos] < minimo, data[Datos] > maximo)
    return mascara_outliers
```

```python
def remove_outliers(df, column_name):
    # Calcula el rango intercuartílico (IQR)
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define los límites inferior y superior para identificar outliers
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Filtra el DataFrame para eliminar outliers
    df_filtered = df[(df[column_name] >= lower_limit) & (df[column_name] <= upper_limit)]
    print(f"{len(df) - len(df_filtered)} ==> '{column_name}' Outliers eliminados")
    return df_filtered
```

```python
def modelPipeline(X_train, X_test, y_train, y_test):
    pipelines = []
    
    # Crear instancias de los clasificadores
    KNN_model = KNeighborsClassifier()
    GaussianNB_model = GaussianNB()
    SVC_model = SVC()
    DecisionTreeClassifier_model = DecisionTreeClassifier()  # Instanciar Support Vector Classifier
    LogisticRegression_model = LogisticRegression()
    RF_model = RandomForestClassifier()
    Gradient_model = GradientBoostingClassifier()
    AdaBOOST_model = AdaBoostClassifier()
    Extre_tree_model = ExtraTreesClassifier()# Instanciar Logistic Regression
  
    
    
    
    cat_features = columns_to_encode

    num_features = columns_to_scale
    #creo pileline para las variables numericas
    numeric_transformer = Pipeline(
                          steps=[("imputer", SimpleImputer(strategy="mean",add_indicator= True)), 
                                 ("scaler", StandardScaler())]
                          )
    #creo pileline para las variables categóricas
    categorical_transformer = Pipeline(
        steps=[
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("encoder", OneHotEncoder(handle_unknown="ignore")),
        ]
    )


    # rear un objeto utilizando la clase ColumnTransformer que nos permite realizar transformaciones en las columnas, que justamente es lo que buscamos
    column_processor = ColumnTransformer(
        transformers=[
            ("numéricas", numeric_transformer, num_features),
            ("categóricas", categorical_transformer, cat_features),
        ]
    )
 

    clfs = [
            ('DecisionTreeClassifier', DecisionTreeClassifier_model), 
            ('RandomForestClassifier', RF_model), 
            ('LogisticRegression', LogisticRegression_model), 
            ("KNN_model" , KNN_model),
            #("GaussianNB_model" , GaussianNB_model),
            ("SVC_model" , SVC_model),
            ("DecisionTreeClassifier_model" , DecisionTreeClassifier_model),  # Instanciar Support Vector Classifier
            ("Gradient_model" , Gradient_model),
            ("AdaBOOST_model" , AdaBOOST_model),
            ("Extre_tree_model" , Extre_tree_model)
            ]


    
    scores_df = pd.DataFrame(columns=['Model', 'accuracy', 'recall',"Precision","f1-score","Tiempo Entrenamiento"])
    

    for clf_name, clf in clfs:
        
        pipeline = Pipeline(steps=[("preprocessor", column_processor), ("Clasificacion", clf)])
        start_train = time.time()
        pipeline.fit(X_train, y_train)
        end_train = time.time()
        y_pred = pipeline.predict(X_test)
        
        acuracy = skm.accuracy_score(y_test, y_pred)
        
        recall = skm.recall_score(y_test, y_pred)
        
        precision = skm.precision_score(y_test, y_pred)
        
        f1= skm.f1_score(y_test, y_pred)
        pipelines.append(f"{pipeline}_{clf_name}")
        
        
        print(f"{clf_name} ok!")
        scores_df = scores_df.append({
                                      'Model' : clf_name, 
                                      'accuracy' : acuracy,
                                      'recall' : recall,
                                      "Precision" : precision,
                                      "f1-score" : f1,
                                      "Tiempo Entrenamiento" : f"{end_train - start_train:.3f}s"
                                      }, 
                                     ignore_index=True)
        
     
    return pipelines, scores_df
```

```python
pipelines, score_df = modelPipeline(X_train_split, X_val, y_train_split, y_val)
```
